# **Create Greeting Card using Matplotlib**

```python
# Import required library
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
# 1. Load the background image
img = Image.open("E:\\Self Learning\\03. Learn Python\\Python_Workspace\\Mixed_Codes\\Mecca.jpg")
```

```python
# 2. Create the plot
fig, ax = plt.subplots(figsize=(10, 10))

# 3. Display the image
# 'extent' defines the coordinates of the image edges [left, right, bottom, top]
# This is crucial so your data points align with the image features

ax.imshow(img, extent=[-20, 20, -20, 15])

# 4. Plot your data on top
t = np.linspace(0, 2 * np.pi, 1000)
x = 16 * np.sin(t)**3
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)

# Chart settings
plt.title("With love because of Allah Ta'ala, I'll give you")
plt.plot(x, y, color='red', linewidth=3)
plt.grid(False)
# plt.xlabel('x')
# plt.ylabel('y')
# plt.axhline(0, color='black', linewidth=0.5)
# plt.axvline(0, color='black', linewidth=0.5)
fig.set_facecolor('turquoise')


# 5. Add text Greeting
name = "Selamat Hari Raya Idul Fitri 1447 H"
plt.text(0, 3, name, fontsize=22, fontname='Segoe UI', weight='bold', color='black', ha='center')
name = "Taqabbalallahu minna wa minkum"
plt.text(0, 1, name, fontsize=20, fontname='Times New Roman', weight='bold', color='black', ha='center')
name = "Minal aidzin wal faidzin"
plt.text(0, -2, name, fontsize=16, fontname='Times New Roman', weight='bold', color='black', ha='center')
name = "Mohon maaf lahir dan batin"
plt.text(0, -3, name, fontsize=16, fontname='Times New Roman', weight='bold', color='black', ha='center')
name = "-Fatwa Rafiudin dan Keluarga-"
plt.text(0, -6, name, fontsize=15, fontname='Times New Roman', weight='bold', color='lime', ha='center')

plt.show()
```
